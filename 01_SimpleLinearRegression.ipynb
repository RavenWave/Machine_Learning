# **4. Spliting the dataset into X and y**

> X: independent variables
>
> y: dependent variable
>
> **y = a * X + b**
>
> a: slope
>
> b: intercept

```python
X = df['total_spend']
y = df['sales']
```

```python
# we can use only for simple linear regression:

np.polyfit(X, y, deg=1)
```

>a = 0.04868788
>
>b = 4.24302822
>
>y = 0.04868788 * X + 4.24302822

## **Determination of coeficcients (slopes and intercept)**

```python
slope, intercept = np.polyfit(X, y, deg=1)
```

```python
print(f"Slope: {slope:>23}")
print(f"Intercept: {intercept:<7}")
```

```python
b1, b0 = np.polyfit(X, y, deg=1) 
```

```python
print("b1 :", b1)
print("b0 :", b0)
```

```python
# formulize the y_pred

y_pred = b1 * X + b0
```

```python
y_pred
```

```python
values = {"actual": y, "predicted":y_pred,"residual": y - y_pred, "LSE": (y - y_pred)**2 }
df_2 = pd.DataFrame(values)
df_2
```

```python
df_2.LSE.sum()
```

# **5. Prediction with simple linear regression**

```python
plt.plot(potential_spend, predicted_sales_lin);
```

```python
plt.plot(potential_spend, predicted_sales_lin)
sns.scatterplot(x="total_spend", y="sales", data =df);
```

# **6. How can we improve our predictions?**

> *We can increase the degree of polynomial.*

```python
a = np.polyfit(X,y, deg =3)
a
```

```python
a1 = np.polyfit(X, y, deg=3)[0]   # coefficient of x^3 
a2 = np.polyfit(X, y, deg=3)[1]   # coefficient of x^2 
a3 = np.polyfit(X, y, deg=3)[2]   # coefficient of x 
a0 = np.polyfit(X, y, deg=3)[3]   # intercept
```

```python
predicted_sales_poly = (a1 * potential_spend ** 3) + (a2 * potential_spend ** 2) + (a3 * potential_spend) + (a0)
predicted_sales_poly
```

```python
plt.plot(potential_spend, predicted_sales_poly, color="red" )
```

```python
plt.plot(potential_spend, predicted_sales_poly, color="red" )
sns.scatterplot(x ="total_spend", y = "sales", data = df);
```

```python
plt.plot(potential_spend, predicted_sales_poly, color = "red")
plt.plot(potential_spend, predicted_sales_lin, color = "blue")
sns.scatterplot(x ="total_spend", y = "sales", data = df);
```
